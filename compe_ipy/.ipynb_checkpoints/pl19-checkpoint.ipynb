```python
# coding: utf-8

# ディープコピーをするため、モジュール copy をインポートする。
import copy

# ここから main

print('############################################################')
print('1. ASSIGNMENT')
print('############################################################')
print('')

# 空のリストを生成して代入。
l = []

# 要素（ここではリスト）を追加する。
# 結果として、リストを要素にもつリストができる。
l.append(['A', 'B'])
l.append(['C', 'D'])

# リスト l を copy_of_l に代入する。
copy_of_l = l

# コピー元のリスト ['C', 'D'] の末尾に、文字列 'Y' を追加する。
l[1].append('Y')
# コピー元のリスト [['A', 'B'], ['C', 'D']] の末尾に、リスト ['Z'] を追加する。
l.append(['Z'])

# コピー元のリストと、コピー先のリストを、それぞれ表示する。
print('l         = ', l)
print('copy_of_l = ', copy_of_l)
print('')

# 関数 id を呼び出すことにより、識別値を表示する例。
print('identity of l            =', id(l))
print('identity of copy_of_l    =', id(copy_of_l))
print('identity of l[0]         =', id(l[0]))
print('identity of copy_of_l[0] =', id(copy_of_l[0]))

print('############################################################')
print('2. SHALLOW COPY')
print('############################################################')
print('')

# 空のリストを生成して代入。
l = []

# 要素（ここではリスト）を追加する。
# 結果として、リストを要素にもつリストができる。
l.append(['A', 'B'])
l.append(['C', 'D'])

# リスト l をシャローコピーしたものを copy_of_l に代入する。
copy_of_l = l.copy()

# コピー元のリスト ['C', 'D'] の末尾に、文字列 'Y' を追加する。
l[1].append('Y')
# コピー元のリスト [['A', 'B'], ['C', 'D']] の末尾に、リスト ['Z'] を追加する。
l.append(['Z'])

# コピー元のリストと、コピー先のリストを、それぞれ表示する。
print('l         = ', l)
print('copy_of_l = ', copy_of_l)
print('')

# 関数 id を呼び出すことにより、識別値を表示する例。
print('identity of l            =', id(l))
print('identity of copy_of_l    =', id(copy_of_l))
print('identity of l[0]         =', id(l[0]))
print('identity of copy_of_l[0] =', id(copy_of_l[0]))

print('############################################################')
print('3. DEEP COPY')
print('############################################################')
print('')

# 空のリストを生成して代入。
l = []

# 要素（ここではリスト）を追加する。
# 結果として、リストを要素にもつリストができる。
l.append(['A', 'B'])
l.append(['C', 'D'])

# リスト l をディープコピーしたものを copy_of_l に代入する。
copy_of_l = copy.deepcopy(l)

# コピー元のリスト ['C', 'D'] の末尾に、文字列 'Y' を追加する。
l[1].append('Y')
# コピー元のリスト [['A', 'B'], ['C', 'D']] の末尾に、リスト ['Z'] を追加する。
l.append(['Z'])

# コピー元のリストと、コピー先のリストを、それぞれ表示する。
print('l         = ', l)
print('copy_of_l = ', copy_of_l)
print('')

# 関数 id を呼び出すことにより、識別値を表示する例。
print('identity of l            =', id(l))
print('identity of copy_of_l    =', id(copy_of_l))
print('identity of l[0]         =', id(l[0]))
print('identity of copy_of_l[0] =', id(copy_of_l[0]))
```

```python
def pFlip(n,l):
    l[:n] = l[:n][::-1] 
    return l

#main

a = [3,2,1,6,5,4,10,9,8,7,12,11]

f = len(a)

while f > 1:
    ind = a.index(f)
    if f - 1 != ind:
        pFlip(f, pFlip(ind+1,a))
        
    f -= 1
print(a)
```
